#FUNCTIONS


```python
import numpy as np

def find_average(numbers):
    return np.mean(numbers)

# Example usage
data = [10, 20, 30, 40, 50]
avg = find_average(data)
print("Average:", avg)
```

```python
def factorial(a):
    result = 1
    if a == 0 or a == 1:
        return 1
    else:
        for i in range(1, a + 1):
            result *= i
        return result

print(factorial(5))  # Output: 120
```
